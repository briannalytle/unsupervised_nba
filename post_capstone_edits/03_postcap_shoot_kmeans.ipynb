```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pickle

from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

# Import dataframe

```python
nba_stats = pd.read_csv('../data/nba_later_eda.csv', index_col = [0])
```

# Label Stats

```python
shooting_stats = ['FGA_2P%', 'FGA_3P%', '2PT%', '2PT_MR%', '3P%',
           'FBPTS%', 'OFFTOV%', 'PAINTPTS%', 'AST_2FGM%', 'UAST_2FGM%',
           'AST_3FGM%', 'UAST_3FGM%', 'FGM_AST%', 'FGM_UAST%'
           ]
```

# KMEANS

```python
X = nba_stats[shooting_stats]
y = nba_stats['Player']

#scale data
sc = StandardScaler()
X_sc = sc.fit_transform(X)

#dimensionality reduction
pca = PCA(n_components = 5)
pca.fit(X_sc)
X_pca = pca.transform(X_sc)
print("Cumulative Explained Variance:", pca.explained_variance_ratio_.sum())

#creating clusters
shooting = KMeans(n_clusters = 3,random_state= 248)
shooting.fit(X_pca)
y_kmeans = shooting.predict(X_pca)
nba_stats['shoot_kmean'] = shooting.labels_
shoot_sil = silhouette_score(X_sc, shooting.labels_)
silhouette_score(X_sc, shooting.labels_)
```

```python
shoot_cluster0 = nba_stats[nba_stats['shoot_kmean'] == 0]
shoot_cluster1 = nba_stats[nba_stats['shoot_kmean'] == 1]
shoot_cluster2 = nba_stats[nba_stats['shoot_kmean'] == 2]
```
