```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pickle

from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

# Import dataframe

```python
nba_stats = pd.read_csv('../data/nba_later_eda.csv', index_col = [0])
```

# Label Stats

```python
defensive_stats = ['DREB', 'DEF_RTG', 'DREB%', 'T_DREB%', 'STL', 'T_STL%',
            'BLK', 'DEF_WS', 'T_BLKA%',
            'BLKA',  'FP']
```

# KMEANS

```python
X = nba_stats[defensive_stats]
y = nba_stats['Player']

#scale data
sc = StandardScaler()
X_sc = sc.fit_transform(X)

#dimensionality reduction
pca = PCA(n_components = 5)
pca.fit(X_sc)
X_pca = pca.transform(X_sc)
print("Cumulative Explained Variance:", pca.explained_variance_ratio_.sum())

#creating clusters
defensive = KMeans(n_clusters = 3, random_state= 248)
defensive.fit(X_pca)
y_kmeans = defensive.predict(X_pca)
nba_stats['def_kmean'] = defensive.labels_
def_sil = silhouette_score(X_sc, defensive.labels_)
print("Silhouette Score:", silhouette_score(X_sc, defensive.labels_))
```

```python
def_cluster0 = nba_stats[nba_stats['def_kmean'] == 0]
def_cluster1 = nba_stats[nba_stats['def_kmean'] == 1]
def_cluster2 = nba_stats[nba_stats['def_kmean'] == 2]
```

# Boxplots

```python
fig, axes = plt.subplots(4, 3, figsize = (15, 9))
plt.subplots_adjust(hspace = 0.8)
sns.boxplot(x = 'DREB%', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[0,0])
sns.boxplot(x = 'T_DREB%', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[0,1])
sns.boxplot(x = 'DEF_RTG', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[0,2])
#---------------------- line 2
sns.boxplot(x = 'STL', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[1,0])
sns.boxplot(x = 'T_STL%', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[1,1])
sns.boxplot(x = 'DEF_WS', y = 'def_kmean', data = nba_stats, orient="h", palette="Set2", ax = axes[1,2])
#---------------------- LINE 3
sns.boxplot(x = 'USG%', y = 'def_kmean', data = nba_stats, orient="h", ax = axes[2,0])
sns.boxplot(x = 'T_BLKA%', y = 'def_kmean', data = nba_stats, orient="h", ax = axes[2,2])
#---------------------- LINE 4
sns.boxplot(x = 'OP_TOV', y = 'def_kmean', data = nba_stats, orient="h", ax = axes[3,0])
sns.boxplot(x = 'OP_2NDPTS', y = 'def_kmean', data = nba_stats, orient="h", ax = axes[3,1])
sns.boxplot(x = 'OP_PAINT', y = 'def_kmean', data = nba_stats, orient="h", ax = axes[3,2])
```
