```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import pickle

from scipy import sparse
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics import silhouette_score
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

# Import DataFrame

```python
nba_stats19 = pd.read_csv('../data/nba_stats19.csv', index_col = [0])
nba_stats19.drop(columns = ['DREB_y', 'T_DREB%_y', 'T_STL%_y', 'T_BLK%_y', 'Age_y','FT%_y'], inplace = True)
nba_stats19.rename(columns = {'Age_x': 'Age', 'FT%_x': 'FT%', 'DREB_x': 'DREB', 'T_DREB%_x' : 'T_DREB%',
                             'T_STL%_x': 'T_STL%', 'T_BLK%_x': 'T_BLK%'}, inplace = True)
```

# Label Stats

```python
overall_stats = ['REB','AST', 'DD2', 'TD3', 'USG%', 'T_REB%',
          'T_AST%', 'PTS_TOV', 'BLK', 'T_BLKA%', 'AST_2FGM%', 'AST_3FGM%', 'STL', 'T_STL%',
                'FG%', 'FGA_2P%', 'T_FGA%', 'TOV', 'T_DREB%', 'T_OREB%', 'MP']
```

```python
nba_stats19 = nba_stats19[nba_stats19['MP'] > 9]
```

# KMEANS

```python
X = nba_stats19[overall_stats]
y = nba_stats19['Player']

#scale data
sc = StandardScaler()
X_sc = sc.fit_transform(X)

#dimensionality reduction
pca = PCA(n_components = 5)
pca.fit(X_sc)
X_pca = pca.transform(X_sc)
print("Cumulative Explained Variance:", pca.explained_variance_ratio_.sum())
```

```python
#creating clusters
overall = KMeans(n_clusters = 3,random_state= 248)
overall.fit(X_pca)
y_kmeans = overall.predict(X_pca)
nba_stats19['overall_kmean'] = overall.labels_
overall_sil = silhouette_score(X_sc, overall.labels_)
silhouette_score(X_sc, overall.labels_)
```

**Cluster 1 - Caretaker:** This player doesn't necessarily have the highest amount of points score; however, they will do anything it takes when their team is in danger. This player has the most fast break points, turnovers and don't need much assistance for their attempted FG. Compared to the other 2 clusters, these players have the lowest spread for FG%. This shows me they will take a shot when needed but won't do it just to get points for themselves, moreso for their team. 

**Cluster 2 - IronMan:** This player will take "superhero" actions - the ones' that would probably end up on BleacherReport's twitter. This player completes most of their team steal's when he is on the court and has the msot success with his block attempts. In addition, Cluster 2 has the highest FG% average among the clusters. This player is most likely surrounded by a strong defensive team because though they have a high amoung of defensive rebounds, they only complete a small amoung of their team's total defensive rebounds when on the court.
